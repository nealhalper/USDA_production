```python
sales_fieldcrops = pd.read_csv("../data/sales_fieldcrops.csv")
```

```python
sales_fieldcrops
```

```python
fieldcrops_areaoperated = sales_fieldcrops.loc[sales_fieldcrops['Domain'] == 'AREA OPERATED']
```

```python
fieldcrops_areaoperated = fieldcrops_areaoperated[['Year', 'Commodity', 'Data Item', 'Domain', 'Domain Category', 'Value']]
fieldcrops_areaoperated = fieldcrops_areaoperated.loc[(fieldcrops_areaoperated['Value'] != ' (D)')]
fieldcrops_areaoperated = fieldcrops_areaoperated.loc[(fieldcrops_areaoperated['Value'] != ' (Z)')]
fieldcrops_areaoperated['Value'] = fieldcrops_areaoperated['Value'].str.replace(',', '')
fieldcrops_areaoperated['Value'] = pd.to_numeric(fieldcrops_areaoperated.Value, downcast='integer').fillna(0)
```

```python
fieldcrops_areaoperated = fieldcrops_areaoperated.groupby(['Year', 'Data Item'])['Value'].sum()
```

```python
fieldcrops_areaoperated = pd.DataFrame(data = fieldcrops_areaoperated).reset_index().rename(columns = {'Value':'Area Operated'})
```

```python
fieldcrops_areaoperated
```

```python
fieldcrops_areaoperated.to_csv(path_or_buf = 'fieldcrops_areaoperated.csv', index = False)
```

```python
plt.figure(figsize = (18,10))
plt.bar(fieldcrops_areaoperated['Year'], fieldcrops_areaoperated['Crops output: Vegetables and melons'])
plt.ylabel('Total Factor Productivity')
plt.title('Total Factor Productivity for Vegetables and melons 1948-2021')
plt.show()
```
